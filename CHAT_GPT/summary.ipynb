```python
import openai, os
from openai import OpenAI
from dotenv import load_dotenv, dotenv_values

import nltk
nltk.download('punkt')

load_dotenv() #generate the environment

openai.api_key = os.getenv('OPENAI_API_KEY') #environment api key

client = OpenAI()

#4097 max tokens (using gpt2 tokenizer)

lecture_content = '''
Title: "Synthesis of Art and Technology: Navigating the Digital Renaissance"
Good day, everyone. Today, we immerse ourselves in the captivating realm where art and technology converge, giving rise to a digital renaissance. The dynamic interplay between artistic expression and cutting-edge technology has ushered in a new era of creativity and innovation.
Exploring the Digital Canvas:
We begin our journey by delving into the transformation of the traditional artist's canvas into a digital landscape. The evolution of digital art tools, from graphic tablets to virtual reality platforms, has empowered artists to explore boundless possibilities and redefine the boundaries of their imagination.
Code as a Medium:
In this digital age, programming languages have become a canvas for a different kind of artist—the coder. We explore how code serves not only as a functional tool but as a medium for artistic expression. From generative art to interactive installations, we witness the fusion of programming prowess and creative vision.
Virtual Reality and Immersive Experiences:
Our exploration extends into the realm of virtual reality, where artists transport audiences into immersive experiences. We discuss the transformative power of VR in art, allowing viewers to step into the artist's creation and become active participants in the narrative.
Artificial Intelligence as a Collaborator:
As we navigate this digital renaissance, we encounter the symbiotic relationship between artists and artificial intelligence. AI algorithms become collaborators, generating unique insights, suggesting creative directions, and challenging traditional notions of authorship in the artistic process.
Challenges and Ethical Considerations:
While the marriage of art and technology presents exciting possibilities, we also address the challenges and ethical considerations that arise. From issues of digital ownership to concerns about the impact of technology on human creativity, we navigate the complexities of this evolving landscape.
The Democratization of Art:
One of the remarkable outcomes of the digital renaissance is the democratization of art. We discuss how technology has enabled artists from diverse backgrounds to share their creations globally, fostering a more inclusive and diverse artistic community.
Future Horizons:
In conclusion, we gaze towards the future horizons of the synthesis of art and technology. What new forms of artistic expression await us? How will emerging technologies shape the way we perceive and create art? Join me on this exploration of the digital frontier, where pixels and creativity intertwine to redefine the very essence of artistic possibility.

Title: "Synthesis of Art and Technology: Navigating the Digital Renaissance"
Good day, everyone. Today, we immerse ourselves in the captivating realm where art and technology converge, giving rise to a digital renaissance. The dynamic interplay between artistic expression and cutting-edge technology has ushered in a new era of creativity and innovation.
Exploring the Digital Canvas:
We begin our journey by delving into the transformation of the traditional artist's canvas into a digital landscape. The evolution of digital art tools, from graphic tablets to virtual reality platforms, has empowered artists to explore boundless possibilities and redefine the boundaries of their imagination.
Code as a Medium:
In this digital age, programming languages have become a canvas for a different kind of artist—the coder. We explore how code serves not only as a functional tool but as a medium for artistic expression. From generative art to interactive installations, we witness the fusion of programming prowess and creative vision.
Virtual Reality and Immersive Experiences:
Our exploration extends into the realm of virtual reality, where artists transport audiences into immersive experiences. We discuss the transformative power of VR in art, allowing viewers to step into the artist's creation and become active participants in the narrative.
Artificial Intelligence as a Collaborator:
As we navigate this digital renaissance, we encounter the symbiotic relationship between artists and artificial intelligence. AI algorithms become collaborators, generating unique insights, suggesting creative directions, and challenging traditional notions of authorship in the artistic process.
Challenges and Ethical Considerations:
While the marriage of art and technology presents exciting possibilities, we also address the challenges and ethical considerations that arise. From issues of digital ownership to concerns about the impact of technology on human creativity, we navigate the complexities of this evolving landscape.
The Democratization of Art:
One of the remarkable outcomes of the digital renaissance is the democratization of art. We discuss how technology has enabled artists from diverse backgrounds to share their creations globally, fostering a more inclusive and diverse artistic community.
Future Horizons:
In conclusion, we gaze towards the future horizons of the synthesis of art and technology. What new forms of artistic expression await us? How will emerging technologies shape the way we perceive and create art? Join me on this exploration of the digital frontier, where pixels and creativity intertwine to redefine the very essence of artistic possibility.


Title: "Synthesis of Art and Technology: Navigating the Digital Renaissance"
Good day, everyone. Today, we immerse ourselves in the captivating realm where art and technology converge, giving rise to a digital renaissance. The dynamic interplay between artistic expression and cutting-edge technology has ushered in a new era of creativity and innovation.
Exploring the Digital Canvas:
We begin our journey by delving into the transformation of the traditional artist's canvas into a digital landscape. The evolution of digital art tools, from graphic tablets to virtual reality platforms, has empowered artists to explore boundless possibilities and redefine the boundaries of their imagination.
Code as a Medium:
In this digital age, programming languages have become a canvas for a different kind of artist—the coder. We explore how code serves not only as a functional tool but as a medium for artistic expression. From generative art to interactive installations, we witness the fusion of programming prowess and creative vision.
Virtual Reality and Immersive Experiences:
Our exploration extends into the realm of virtual reality, where artists transport audiences into immersive experiences. We discuss the transformative power of VR in art, allowing viewers to step into the artist's creation and become active participants in the narrative.
Artificial Intelligence as a Collaborator:
As we navigate this digital renaissance, we encounter the symbiotic relationship between artists and artificial intelligence. AI algorithms become collaborators, generating unique insights, suggesting creative directions, and challenging traditional notions of authorship in the artistic process.
Challenges and Ethical Considerations:
While the marriage of art and technology presents exciting possibilities, we also address the challenges and ethical considerations that arise. From issues of digital ownership to concerns about the impact of technology on human creativity, we navigate the complexities of this evolving landscape.
The Democratization of Art:
One of the remarkable outcomes of the digital renaissance is the democratization of art. We discuss how technology has enabled artists from diverse backgrounds to share their creations globally, fostering a more inclusive and diverse artistic community.
Future Horizons:
In conclusion, we gaze towards the future horizons of the synthesis of art and technology. What new forms of artistic expression await us? How will emerging technologies shape the way we perceive and create art? Join me on this exploration of the digital frontier, where pixels and creativity intertwine to redefine the very essence of artistic possibility.

Title: "Synthesis of Art and Technology: Navigating the Digital Renaissance"
Good day, everyone. Today, we immerse ourselves in the captivating realm where art and technology converge, giving rise to a digital renaissance. The dynamic interplay between artistic expression and cutting-edge technology has ushered in a new era of creativity and innovation.
Exploring the Digital Canvas:
We begin our journey by delving into the transformation of the traditional artist's canvas into a digital landscape. The evolution of digital art tools, from graphic tablets to virtual reality platforms, has empowered artists to explore boundless possibilities and redefine the boundaries of their imagination.
Code as a Medium:
In this digital age, programming languages have become a canvas for a different kind of artist—the coder. We explore how code serves not only as a functional tool but as a medium for artistic expression. From generative art to interactive installations, we witness the fusion of programming prowess and creative vision.
Virtual Reality and Immersive Experiences:
Our exploration extends into the realm of virtual reality, where artists transport audiences into immersive experiences. We discuss the transformative power of VR in art, allowing viewers to step into the artist's creation and become active participants in the narrative.
Artificial Intelligence as a Collaborator:
As we navigate this digital renaissance, we encounter the symbiotic relationship between artists and artificial intelligence. AI algorithms become collaborators, generating unique insights, suggesting creative directions, and challenging traditional notions of authorship in the artistic process.
Challenges and Ethical Considerations:
While the marriage of art and technology presents exciting possibilities, we also address the challenges and ethical considerations that arise. From issues of digital ownership to concerns about the impact of technology on human creativity, we navigate the complexities of this evolving landscape.
The Democratization of Art:
One of the remarkable outcomes of the digital renaissance is the democratization of art. We discuss how technology has enabled artists from diverse backgrounds to share their creations globally, fostering a more inclusive and diverse artistic community.
Future Horizons:
In conclusion, we gaze towards the future horizons of the synthesis of art and technology. What new forms of artistic expression await us? How will emerging technologies shape the way we perceive and create art? Join me on this exploration of the digital frontier, where pixels and creativity intertwine to redefine the very essence of artistic possibility.



Title: "Synthesis of Art and Technology: Navigating the Digital Renaissance"
Good day, everyone. Today, we immerse ourselves in the captivating realm where art and technology converge, giving rise to a digital renaissance. The dynamic interplay between artistic expression and cutting-edge technology has ushered in a new era of creativity and innovation.
Exploring the Digital Canvas:
We begin our journey by delving into the transformation of the traditional artist's canvas into a digital landscape. The evolution of digital art tools, from graphic tablets to virtual reality platforms, has empowered artists to explore boundless possibilities and redefine the boundaries of their imagination.
Code as a Medium:
In this digital age, programming languages have become a canvas for a different kind of artist—the coder. We explore how code serves not only as a functional tool but as a medium for artistic expression. From generative art to interactive installations, we witness the fusion of programming prowess and creative vision.
Virtual Reality and Immersive Experiences:
Our exploration extends into the realm of virtual reality, where artists transport audiences into immersive experiences. We discuss the transformative power of VR in art, allowing viewers to step into the artist's creation and become active participants in the narrative.
Artificial Intelligence as a Collaborator:
As we navigate this digital renaissance, we encounter the symbiotic relationship between artists and artificial intelligence. AI algorithms become collaborators, generating unique insights, suggesting creative directions, and challenging traditional notions of authorship in the artistic process.
Challenges and Ethical Considerations:
While the marriage of art and technology presents exciting possibilities, we also address the challenges and ethical considerations that arise. From issues of digital ownership to concerns about the impact of technology on human creativity, we navigate the complexities of this evolving landscape.
The Democratization of Art:
One of the remarkable outcomes of the digital renaissance is the democratization of art. We discuss how technology has enabled artists from diverse backgrounds to share their creations globally, fostering a more inclusive and diverse artistic community.
Future Horizons:
In conclusion, we gaze towards the future horizons of the synthesis of art and technology. What new forms of artistic expression await us? How will emerging technologies shape the way we perceive and create art? Join me on this exploration of the digital frontier, where pixels and creativity intertwine to redefine the very essence of artistic possibility.

Title: "Synthesis of Art and Technology: Navigating the Digital Renaissance"
Good day, everyone. Today, we immerse ourselves in the captivating realm where art and technology converge, giving rise to a digital renaissance. The dynamic interplay between artistic expression and cutting-edge technology has ushered in a new era of creativity and innovation.
Exploring the Digital Canvas:
We begin our journey by delving into the transformation of the traditional artist's canvas into a digital landscape. The evolution of digital art tools, from graphic tablets to virtual reality platforms, has empowered artists to explore boundless possibilities and redefine the boundaries of their imagination.
Code as a Medium:
In this digital age, programming languages have become a canvas for a different kind of artist—the coder. We explore how code serves not only as a functional tool but as a medium for artistic expression. From generative art to interactive installations, we witness the fusion of programming prowess and creative vision.
Virtual Reality and Immersive Experiences:
Our exploration extends into the realm of virtual reality, where artists transport audiences into immersive experiences. We discuss the transformative power of VR in art, allowing viewers to step into the artist's creation and become active participants in the narrative.
Artificial Intelligence as a Collaborator:
As we navigate this digital renaissance, we encounter the symbiotic relationship between artists and artificial intelligence. AI algorithms become collaborators, generating unique insights, suggesting creative directions, and challenging traditional notions of authorship in the artistic process.
Challenges and Ethical Considerations:
While the marriage of art and technology presents exciting possibilities, we also address the challenges and ethical considerations that arise. From issues of digital ownership to concerns about the impact of technology on human creativity, we navigate the complexities of this evolving landscape.
The Democratization of Art:
One of the remarkable outcomes of the digital renaissance is the democratization of art. We discuss how technology has enabled artists from diverse backgrounds to share their creations globally, fostering a more inclusive and diverse artistic community.
Future Horizons:
In conclusion, we gaze towards the future horizons of the synthesis of art and technology. What new forms of artistic expression await us? How will emerging technologies shape the way we perceive and create art? Join me on this exploration of the digital frontier, where pixels and creativity intertwine to redefine the very essence of artistic possibility.


Title: "Synthesis of Art and Technology: Navigating the Digital Renaissance"
Good day, everyone. Today, we immerse ourselves in the captivating realm where art and technology converge, giving rise to a digital renaissance. The dynamic interplay between artistic expression and cutting-edge technology has ushered in a new era of creativity and innovation.
Exploring the Digital Canvas:
We begin our journey by delving into the transformation of the traditional artist's canvas into a digital landscape. The evolution of digital art tools, from graphic tablets to virtual reality platforms, has empowered artists to explore boundless possibilities and redefine the boundaries of their imagination.
Code as a Medium:
In this digital age, programming languages have become a canvas for a different kind of artist—the coder. We explore how code serves not only as a functional tool but as a medium for artistic expression. From generative art to interactive installations, we witness the fusion of programming prowess and creative vision.
Virtual Reality and Immersive Experiences:
Our exploration extends into the realm of virtual reality, where artists transport audiences into immersive experiences. We discuss the transformative power of VR in art, allowing viewers to step into the artist's creation and become active participants in the narrative.
Artificial Intelligence as a Collaborator:
As we navigate this digital renaissance, we encounter the symbiotic relationship between artists and artificial intelligence. AI algorithms become collaborators, generating unique insights, suggesting creative directions, and challenging traditional notions of authorship in the artistic process.
Challenges and Ethical Considerations:
While the marriage of art and technology presents exciting possibilities, we also address the challenges and ethical considerations that arise. From issues of digital ownership to concerns about the impact of technology on human creativity, we navigate the complexities of this evolving landscape.
The Democratization of Art:
One of the remarkable outcomes of the digital renaissance is the democratization of art. We discuss how technology has enabled artists from diverse backgrounds to share their creations globally, fostering a more inclusive and diverse artistic community.
Future Horizons:
In conclusion, we gaze towards the future horizons of the synthesis of art and technology. What new forms of artistic expression await us? How will emerging technologies shape the way we perceive and create art? Join me on this exploration of the digital frontier, where pixels and creativity intertwine to redefine the very essence of artistic possibility.

Title: "Synthesis of Art and Technology: Navigating the Digital Renaissance"
Good day, everyone. Today, we immerse ourselves in the captivating realm where art and technology converge, giving rise to a digital renaissance. The dynamic interplay between artistic expression and cutting-edge technology has ushered in a new era of creativity and innovation.
Exploring the Digital Canvas:
We begin our journey by delving into the transformation of the traditional artist's canvas into a digital landscape. The evolution of digital art tools, from graphic tablets to virtual reality platforms, has empowered artists to explore boundless possibilities and redefine the boundaries of their imagination.
Code as a Medium:
In this digital age, programming languages have become a canvas for a different kind of artist—the coder. We explore how code serves not only as a functional tool but as a medium for artistic expression. From generative art to interactive installations, we witness the fusion of programming prowess and creative vision.
Virtual Reality and Immersive Experiences:
Our exploration extends into the realm of virtual reality, where artists transport audiences into immersive experiences. We discuss the transformative power of VR in art, allowing viewers to step into the artist's creation and become active participants in the narrative.
Artificial Intelligence as a Collaborator:
As we navigate this digital renaissance, we encounter the symbiotic relationship between artists and artificial intelligence. AI algorithms become collaborators, generating unique insights, suggesting creative directions, and challenging traditional notions of authorship in the artistic process.
Challenges and Ethical Considerations:
While the marriage of art and technology presents exciting possibilities, we also address the challenges and ethical considerations that arise. From issues of digital ownership to concerns about the impact of technology on human creativity, we navigate the complexities of this evolving landscape.
The Democratization of Art:
One of the remarkable outcomes of the digital renaissance is the democratization of art. We discuss how technology has enabled artists from diverse backgrounds to share their creations globally, fostering a more inclusive and diverse artistic community.
Future Horizons:
In conclusion, we gaze towards the future horizons of the synthesis of art and technology. What new forms of artistic expression await us? How will emerging technologies shape the way we perceive and create art? Join me on this exploration of the digital frontier, where pixels and creativity intertwine to redefine the very essence of artistic possibility.




Title: "Synthesis of Art and Technology: Navigating the Digital Renaissance"
Good day, everyone. Today, we immerse ourselves in the captivating realm where art and technology converge, giving rise to a digital renaissance. The dynamic interplay between artistic expression and cutting-edge technology has ushered in a new era of creativity and innovation.
Exploring the Digital Canvas:
We begin our journey by delving into the transformation of the traditional artist's canvas into a digital landscape. The evolution of digital art tools, from graphic tablets to virtual reality platforms, has empowered artists to explore boundless possibilities and redefine the boundaries of their imagination.
Code as a Medium:
In this digital age, programming languages have become a canvas for a different kind of artist—the coder. We explore how code serves not only as a functional tool but as a medium for artistic expression. From generative art to interactive installations, we witness the fusion of programming prowess and creative vision.
Virtual Reality and Immersive Experiences:
Our exploration extends into the realm of virtual reality, where artists transport audiences into immersive experiences. We discuss the transformative power of VR in art, allowing viewers to step into the artist's creation and become active participants in the narrative.
Artificial Intelligence as a Collaborator:
As we navigate this digital renaissance, we encounter the symbiotic relationship between artists and artificial intelligence. AI algorithms become collaborators, generating unique insights, suggesting creative directions, and challenging traditional notions of authorship in the artistic process.
Challenges and Ethical Considerations:
While the marriage of art and technology presents exciting possibilities, we also address the challenges and ethical considerations that arise. From issues of digital ownership to concerns about the impact of technology on human creativity, we navigate the complexities of this evolving landscape.
emarkable outcomes of the digital renaissance is the democratization of art. We discuss how technology has enabled artists from diverse backgrounds to share their creations globally, fostering a more inclusive and diverse artistic community.
Future Horizons:
In conclusion, we gaze towards the future horizons of the synthesis of art and technology. What new forms of artistic expression await us? How will emerging technologies shape the way we perceive and create art? Join me on this exploration of the digital frontier, where pixels and creativity intertwine to redefine the very essence of artistic possibility.
'''
#receive response
def get_completion(lecture_content):
    prompt = "Can you summarize the main ideas of this lecture:" + lecture_content

    tokens = nltk.word_tokenize(prompt)

    if len(tokens) > 3482: #splitting up the lecture length into different length and summarise it separately
        prompt_lst = []
        if len(tokens) % 3482 != 0:
            times = len(tokens) // 3482 + 1
        else:
            times = len(tokens) / 3482
        start = 0
        end = 3482
        for i in range(times):
            prompt_lst.append(prompt[start: end])
            start += 3482
            end += 3482
        summary = ''
        for i in range(0, times):
            response = openai.completions.create(
            model="gpt-3.5-turbo-1106",
#   response_format={ "type": "json_object" },
#   messages=[
#     {"role": "system", "content": "You are a helpful assistant designed to output JSON."},
#     {"role": "user", "content": "Who won the world series in 2020?"}
#   ]
# )
# print(response.choices[0].message.content)
            prompt = prompt_lst[i],
            max_tokens = 500
            )
            summary += response.choices[0].text.strip()
            summary += "HERE IS THE BREAK"
        return summary
    else:
        response = openai.client.completion.create(
        engine = "text-davinci-003",
        prompt = prompt,
        max_tokens = 500
        )
        return response.choices[0].text.strip()

#the output
response = get_completion(lecture_content)
print(response)
```
